```python
#   Evaluacion estandar - USAR ESTE
#   Grafica dos variables
import gymnasium   #
import CoppeliaSim_Gym   # Entorno personalizado
import stable_baselines3   # Biblioteca RL
from stable_baselines3 import A2C, PPO
from matplotlib import pyplot as plt 
import pandas as pd
import numpy as np
import seaborn as sns
# Inicializar el entorno
ENV_NAME = "CoppeliaSim_Gym/GymCoppManR-v0"
env = gymnasium.make(ENV_NAME)

model = A2C.load("best_model", env=env)
model.policy
ep_reward = 0
# Evaluar el modelo entrenado
vald = []
valr = []
X_ = []
Y1 = []
Y2 = []
obs, inf = env.reset()
print('obs', obs.shape)
for episode in range(100):
    env.render()
    action, state = model.predict(obs, deterministic=False) # ,deterministic=True
    n_obs, reward, done, tru, info = env.step(action)
    #print(f'Obs: {obs} reward: {reward} done: {done} info: {info}')
    print(f'Acciones: {action} Estado: {state} Recompensa: {reward} Info: {info}')
    dis = env.distance_to_goal()
    obs = n_obs
    valr += [reward]
    vald += [dis]
    X_.append(episode)
    Y1.append(np.array(valr).mean())
    Y2.append(np.array(vald).mean())
       
fig, ax1 = plt.subplots(figsize=(7,5)) 
color1 = 'tab:red'
ax1.set_xlabel('Episode')
ax1.set_ylabel('Reward', color=color1)
ax1.plot(X_, Y1, color=color1)
ax1.tick_params(axis='y', labelcolor=color1)
ax2 = ax1.twinx()
color2 = 'tab:blue'
ax2.set_ylabel('Distance (m)', color=color2)
ax2.plot(X_, Y2, color=color2)
ax2.tick_params(axis='y', labelcolor=color2)
plt.title('Test Model')

ax1.grid(color='b', ls = '-.', lw = 0.25)
```

```python
# Evaluacion con ciclo anidado, for - while
import gymnasium   #
import CoppeliaSim_Gym   # Entorno personalizado
import stable_baselines3   # Biblioteca RL
from stable_baselines3 import A2C, PPO
from matplotlib import pyplot as plt 
import pandas as pd
import numpy as np

# Inicializar el entorno
ENV_NAME = "CoppeliaSim_Gym/GymCoppManR-v0"
env = gymnasium.make(ENV_NAME)
model = A2C.load("modeloA2C", env=env)
model.policy
val = []
X_ = []
Y_ = []
all_episode_rewards = []
num_episodes = 25
for i in range(num_episodes):
    episode_rewards = []
    done = False
    obs, info = env.reset()
    while not done:
        # states solo son útiles cuando se usan políticas LSTM
        action, states = model.predict(obs, deterministic=False)
        # here, action, rewards and dones are arrays
        # because we are using vectorized env
        obs, reward, done, tru, info = env.step(action)
        episode_rewards.append(reward)
        val += [reward]
        X_.append(num_episodes)
        Y_.append(np.array(val).mean())
        
    all_episode_rewards.append(sum(episode_rewards))
plt.plot(X_, Y_)
plt.title('Acumulated reward')
plt.xlabel('Episodes')
plt.ylabel('Reward')
plt.show()
mean_episode_reward = np.mean(all_episode_rewards)
print("Mean reward:", mean_episode_reward, "Num episodes:", num_episodes)
env.close()
```
