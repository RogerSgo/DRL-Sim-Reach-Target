```python
import gymnasium   #
import CoppeliaSim_Gym   # Entorno personalizado
import stable_baselines3   # Biblioteca RL
from stable_baselines3 import A2C, PPO
from matplotlib import pyplot as plt 
import pandas as pd
import numpy as np

# Inicializar el entorno
ENV_NAME = "CoppeliaSim_Gym/GymCoppManR-v0"
env = gymnasium.make(ENV_NAME)
model = A2C.load("modeloA2C", env=env)
model.policy
val = []
X_ = []
Y_ = []
all_episode_rewards = []
num_episodes = 25
for i in range(num_episodes):
    episode_rewards = []
    done = False
    obs, info = env.reset()
    while (num_episodes==10):
        # states solo son tiles cuando se usan polticas LSTM
        action, states = model.predict(obs, deterministic=False)
        # here, action, rewards and dones are arrays
        # because we are using vectorized env
        obs, reward, done, tru, info = env.step(action)
        episode_rewards.append(reward)
        val += [reward]
        X_.append(num_episodes)
        Y_.append(np.array(val).mean())
        
    all_episode_rewards.append(sum(episode_rewards))
plt.plot(X_, Y_)
plt.title('Acumulated reward')
plt.xlabel('Episodes')
plt.ylabel('Reward')
plt.show()
mean_episode_reward = np.mean(all_episode_rewards)
print("Mean reward:", mean_episode_reward, "Num episodes:", num_episodes)
env.close()
```
